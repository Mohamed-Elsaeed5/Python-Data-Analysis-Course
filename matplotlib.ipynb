### ðŸ“Š Matplotlib Chart Types and Their Use Cases

| Chart Type            | Use Case                                                                 |
|------------------------|--------------------------------------------------------------------------|
| **Line Plot**          | Show trends or changes over time                                        |
| **Bar Chart**          | Compare quantities between categories                                   |
| **Horizontal Bar**     | Compare quantities horizontally, useful for long labels                 |
| **Stacked Bar**        | Show composition of categories over a total                            |
| **Pie Chart**          | Show percentage breakdown of a whole                                    |
| **Scatter Plot**       | Visualize correlation or distribution between two variables             |
| **Histogram**          | Display frequency distribution of a single numeric variable             |
| **Box Plot**           | Show summary statistics and detect outliers                             |
| **Area Plot**          | Display cumulative trends over categories                               |
| **Multiple Line Plot** | Compare multiple series in the same chart                               |
| **Filled Line Plot**   | Highlight area under a curve (like a cumulative view)                   |
| **Subplots**           | Show multiple charts in one figure for side-by-side comparison          |

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [23, 45, 12, 36],
    'Growth': [5, 15, -2, 7]
}
df = pd.DataFrame(data)

# 1. Line Plot
plt.figure()
plt.plot(df['Category'], df['Values'], marker='o')
plt.title("Line Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
# 2. Bar Chart
plt.figure()
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
# 3. Horizontal Bar Chart
plt.figure()
plt.barh(df['Category'], df['Values'], color='orange')
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Category")
plt.show()
```

```python
# 4. Stacked Bar Chart
plt.figure()
bottom_bar = df['Values'] - df['Growth']
plt.bar(df['Category'], bottom_bar, label='Base')
plt.bar(df['Category'], df['Growth'], bottom=bottom_bar, label='Growth')
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

```python
# 5. Pie Chart
plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
# 6. Scatter Plot
plt.figure()
plt.scatter(df['Values'], df['Growth'], c='red')
plt.title("Scatter Plot")
plt.xlabel("Values")
plt.ylabel("Growth")
plt.grid(True)
plt.show()
```

```python
# 7. Histogram
numeric_data = [23, 45, 12, 36, 45, 23, 23, 12, 45, 36, 36]
plt.figure()
plt.hist(numeric_data, bins=5, color='green', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
# 8. Boxplot
plt.figure()
plt.boxplot(df['Values'])
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```

```python
# 9. Area Plot
plt.figure()
plt.stackplot(df['Category'], df['Values'], df['Growth'], labels=['Values', 'Growth'])
plt.title("Area Plot")
plt.legend()
plt.show()
```

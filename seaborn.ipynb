### ðŸŒŠ Seaborn Chart Types and Use Cases

| Seaborn Chart Type       | Function              | Use Case                                                                 |
|---------------------------|------------------------|--------------------------------------------------------------------------|
| **Bar Plot**              | `sns.barplot()`        | Compare mean values across categories (with optional error bars)         |
| **Count Plot**            | `sns.countplot()`      | Count occurrences of categories                                          |
| **Box Plot**              | `sns.boxplot()`        | Show distributions with medians, quartiles, and outliers                 |
| **Violin Plot**           | `sns.violinplot()`     | Combination of boxplot and KDE to show distribution shape                |
| **Strip Plot**            | `sns.stripplot()`      | Plot individual data points (with jitter for visibility)                 |
| **Swarm Plot**            | `sns.swarmplot()`      | Like stripplot but non-overlapping points                               |
| **Histogram**             | `sns.histplot()`       | Show distribution of single variable                                     |
| **KDE Plot**              | `sns.kdeplot()`        | Smooth distribution curve (Kernel Density Estimate)                      |
| **Dist Plot (deprecated)**| `sns.distplot()`       | Combines histogram + KDE (use histplot + kdeplot now)                   |
| **Scatter Plot**          | `sns.scatterplot()`    | Visualize relationship between two numerical variables                   |
| **Line Plot**             | `sns.lineplot()`       | Show trends over time or continuous variable                             |
| **Joint Plot**            | `sns.jointplot()`      | Combine scatter + hist + correlation in one view                         |
| **Pair Plot**             | `sns.pairplot()`       | Plot pairwise relationships in dataset                                   |
| **Heatmap**               | `sns.heatmap()`        | Correlation matrix or pivot table visualization                          |
| **Clustermap**            | `sns.clustermap()`     | Heatmap + hierarchical clustering                                        |
| **FacetGrid**             | `sns.FacetGrid()`      | Multi-plot grid by categorical variable                                  |
| **Catplot**               | `sns.catplot()`        | Wrapper to plot multiple categorical plots in one interface              |
| **Boxen Plot**            | `sns.boxenplot()`      | Better for large datasets with many outliers                            |
| **Regression Plot**       | `sns.regplot()`        | Scatter plot with linear regression line                                 |
| **LM Plot**               | `sns.lmplot()`         | Same as regplot but supports FacetGrid                                   |


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create custom DataFrame similar to tips
np.random.seed(0)
n = 200
df = pd.DataFrame({
    "total_bill": np.round(np.random.uniform(10, 50, n), 2),
    "tip": np.round(np.random.uniform(1, 10, n), 2),
    "sex": np.random.choice(["Male", "Female"], n),
    "smoker": np.random.choice(["Yes", "No"], n),
    "day": np.random.choice(["Thur", "Fri", "Sat", "Sun"], n),
    "time": np.random.choice(["Lunch", "Dinner"], n),
    "size": np.random.randint(1, 6, n)
})

# 1. Bar Plot
sns.barplot(x="day", y="total_bill", data=df)
plt.title("Bar Plot: Avg Total Bill by Day")
plt.show()
```

```python
# 2. Count Plot
sns.countplot(x="day", data=df)
plt.title("Count Plot: Count by Day")
plt.show()
```

```python
# 3. Box Plot
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Box Plot: Bill by Day")
plt.show()
```

```python
# 4. Violin Plot
sns.violinplot(x="day", y="total_bill", data=df)
plt.title("Violin Plot: Bill by Day")
plt.show()
```

```python
# 5. Strip Plot
sns.stripplot(x="day", y="total_bill", data=df, jitter=True)
plt.title("Strip Plot: Bill by Day")
plt.show()
```

```python
# 6. Swarm Plot
sns.swarmplot(x="day", y="total_bill", data=df)
plt.title("Swarm Plot: Bill by Day")
plt.show()
```

```python
# 7. Histogram
sns.histplot(df["total_bill"], bins=10)
plt.title("Histogram: Total Bill")
plt.show()
```

```python
# 8. KDE Plot (fixed for recent seaborn versions)
sns.kdeplot(x=df["total_bill"], fill=True)
plt.title("KDE Plot: Total Bill Density")
plt.show()
```

```python
# 9. Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Scatter: Bill vs Tip")
plt.show()
```

```python
# 10. Line Plot
sns.lineplot(x="size", y="tip", data=df)
plt.title("Line Plot: Tip by Party Size")
plt.show()
```

```python
# 11. Joint Plot
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter")
plt.suptitle("Joint Plot: Tip vs Bill", y=1.02)
plt.show()
```

```python
# 12. Pair Plot
sns.pairplot(df[["total_bill", "tip", "size"]])
plt.suptitle("Pair Plot: Numeric Columns", y=1.02)
plt.show()
```

```python
# 13. Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap: Correlation")
plt.show()
```

```python
# 14. Clustermap
sns.clustermap(corr, annot=True, cmap="viridis")
plt.title("Clustermap")
plt.show()
```

```python
# 15. FacetGrid
g = sns.FacetGrid(df, col="sex")
g.map_dataframe(sns.histplot, x="total_bill")
g.fig.suptitle("FacetGrid: Bill by Sex", y=1.05)
plt.show()
```

```python
# 16. Catplot
sns.catplot(x="day", y="total_bill", kind="box", data=df)
plt.title("Catplot: Bill by Day")
plt.show()
```

```python
# 17. Boxen Plot
sns.boxenplot(x="day", y="total_bill", data=df)
plt.title("Boxen Plot: Bill by Day")
plt.show()
```

```python
# 18. Regression Plot
sns.regplot(x="total_bill", y="tip", data=df)
plt.title("Regression Plot: Tip vs Bill")
plt.show()
```
